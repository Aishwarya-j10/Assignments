### **Task 7:** Implement a basic linear regression model. 

```python
#- Allow users to select a target variable and one or more predictor variables.
# - Train a linear regression model and display the regression coefficients and R-squared value.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
np.random.seed(0) 
samples = 100 
X = np.random.rand(samples, 2) #TWO PREDICTOR VARIABLES 
Y = 2 * X[:, 0] + 3 * X[:, 1] +np.random.rand(samples) 
df = pd.DataFrame({'Predictor1': X[:, 0], 'Predictor2': X[:, 1], 'Target': Y})
target_variable = 'Target'

# Allow users to select target and predictor variables
print("Available variables:", df.columns)
#target_variable = input("Select the target variable: ")
#predictor_variables = input("Select the predictor variables (comma-separated): ").split(',')
predictor_variables = ['Predictor1', 'Predictor2']
X_train, X_test, y_train, y_test = train_test_split(df[predictor_variables], 
                                                    df[target_variable], 
                                                    test_size=0.2, 
                                                    random_state=42) 
```

```python
model = LinearRegression()


model.fit(X_train, y_train)


y_pred = model.predict(X_test)
r2 = r2_score(y_test, y_pred)
print(f"R-squared value: {r2}")

mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse}")

print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)
```

### **Task 8:** Visualize the regression results. 

```python
# - Plot the actual vs. predicted values.
#- Display the regression line on a scatter plot. 
# Actual vs. Predicted values

plt.scatter(y_test, y_pred)
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title("Actual vs. Predicted")
plt.show()
# Regression line on a scatter plot
#plt.subplot(1, 2, 2)
plt.scatter(X_test[predictor_variables[0]], y_test, color='blue', label='Actual')
plt.scatter(X_test[predictor_variables[0]], y_pred, color='red', label='Predicted')
plt.plot(X_test[predictor_variables[0]], y_pred, color='green', label='Regression Line')
plt.xlabel(predictor_variables[0])
plt.ylabel(target_variable)

plt.title("Regression Line on Scatter Plot")


plt.show()
```
